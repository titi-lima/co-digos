```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definimos o sistema de EDO na forma y' = f(t,y)
# Aqui: y[0] = y(t), y[1] = y'(t)
# Portanto:
# y'[0] = y'(t) = y[1]
# y'[1] = y''(t) = cos(t) - y(t)
def ode(t, Y):
    return [Y[1], np.cos(t) - Y[0]]

# Intervalo de integrao e pontos de avaliao
t_span = (0, 10)
t_eval = np.linspace(0, 10, 200)

# Condies iniciais: y(0)=0, y'(0)=0
y0 = [0, 0]

# Resolvendo numericamente
sol = solve_ivp(ode, t_span, y0, t_eval=t_eval)

# Soluo analtica: y(t) = (1/2) * t * sin(t)
def y_analytic(t):
    return 0.5 * t * np.sin(t)

# Plotando ambas as solues
plt.figure(figsize=(8,5))
plt.plot(sol.t, sol.y[0], 'b-', label='Soluo Numrica')
plt.plot(sol.t, y_analytic(sol.t), 'r--', label='Soluo Analtica')
plt.title('Comparao: y" + y = cos(t), y(0)=0, y\'(0)=0')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.show()
```
